### Decorators in Python - Step by Step Guide

# 1. Understanding Decorators

## 1.1 What is a Decorator?
A decorator is a function that takes another function and extends its behavior without explicitly modifying it.  
They are often used for logging, enforcing access control, instrumentation, caching, etc.

```python
# Decorator function
def my_decorator(func):
    def wrapper():
        print("Something before the function runs.")
        func() # Calls the decorated function
        print("Something after the function runs.")
    return wrapper

@my_decorator # Applying the decorator
def say_hello():
    print("Hello!")

say_hello()
# Output:
# Something before the function runs.
# Hello!
# Something after the function runs.
```

# 2. Decorators with Arguments

```python
# Decorator that takes an argument
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n): # Run the function 'n' times
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def greet(name):
    print(f"Hello, {name}!")

greet("Alice") # Output: "Hello, Alice!" printed 3 times
```

# 3. Practical Use

## 3.1 Timing a Function

```python
import time

# Decorator to measure function execution time
def timing_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Execution Time: {end_time - start_time:.4f} seconds")
        return result
    return wrapper

@timing_decorator
def slow_function():
    time.sleep(2)
    print("Finished slow function")

slow_function()
```

# 4. Chaining Multiple Decorators

```python
def uppercase_decorator(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return wrapper
 
def exclaim_decorator(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs) + "!!!"
    return wrapper

@uppercase_decorator
@exclaim_decorator
def greet_person(name):
    return f"Hello, {name}"

print(greet_person("Alice"))  # Output: HELLO, ALICE!!!
```

# 5. Class-based Decorators

```python
class LoggerDecorator:
    def __init__(self, func):
        self.func = func
        
    def __call__(self, *args, **kwargs):
        print(f"Calling {self.func.__name__} with {args} {kwargs}")
        return self.func(*args, **kwargs)

@LoggerDecorator
def add(a, b):
    return a + b

print(add(3, 4))  # Output: Calling add with (3, 4) {} -> 7
```

# 6. Built-in Decorators (`staticmethod`, `classmethod`, `property`)

```python
class Sample:
    def __init__(self, value):
        self._value = value

    # A method that belongs to the class, but does not access any instance or class variables
    @staticmethod 
    def static_method():
        print("I am a static method")
    
    # A method that receives the class (`cls`) as an implicit first argument and can modify class-level attributes
    @classmethod 
    def class_method(cls):
        print("I am a class method")
    
    # A built-in decorator that allows us to define a method that can be accessed like an attribute
    @property
    def value(self):
        return self._value

s = Sample(10)
s.static_method() # Calling static method
s.class_method() # Calling class method
print(s.value) # Accessing property without parentheses
```
