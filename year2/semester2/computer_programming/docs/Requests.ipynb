## 2. Making a Simple GET Request

We start with a basic GET request to an API (GitHub's public API).


````python
response = requests.get('https://api.github.com')

print("Status Code:", response.status_code)
print("Response Text (partial):", response.text[:200])

# **Explanation**:
# - `requests.get(url)` sends a GET request to the specified URL.
# - `response.status_code` gives the HTTP status code (200, 404, etc.).
# - `response.text` is the raw textual content of the response.
````

## 3. Parsing JSON Responses

Many APIs return data in JSON format. Convert it into Python objects:


````python
response = requests.get('https://api.github.com')
data = response.json()

print("Keys in JSON response:", list(data.keys()))
print("\nFull JSON data:\n", data)

# **Explanation**:
# - `.json()` parses JSON content into a Python dictionary (or list).
# - You can then inspect specific keys (e.g., `data['current_user_url']`).
````

## 4. Using Query Parameters

Some APIs require query parameters. Let's search GitHub repositories for \"python requests\" and sort by stars.


````python
params = {
    'q': 'python requests',
    'sort': 'stars'
}

response = requests.get(
    'https://api.github.com/search/repositories',
    params=params
)

print("Final URL:", response.url)
search_data = response.json()
print("Number of repositories found:", search_data.get('total_count', 'N/A'))

# **Explanation**:
# - We pass a dictionary `params` to `requests.get()`.
# - `response.url` shows the final URL, including `?q=python+requests&sort=stars`.
# - The JSON response often contains a 'total_count' key indicating how many search results there are.
````

## 5. Sending a POST Request

Here's an example of sending data in JSON format. We use \"jsonplaceholder.typicode.com\" to simulate a real POST request.


````python
json_payload = {
    "title": "Hello World",
    "body": "This is a test.",
    "userId": 1
}

response = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=json_payload
)

print("Status Code:", response.status_code)
print("Response JSON:", response.json())

# **Explanation**:
# - `requests.post(url, json=...)` sets `Content-Type: application/json`.
# - The mock API returns a JSON response with an assigned ID, letting us pretend we've created a resource.
````

## 6. Handling Errors and Exceptions

Use `raise_for_status()` and a try-except block to handle bad status codes.


````python
try:
    bad_response = requests.get("https://api.github.com/nonexistent-endpoint")
    bad_response.raise_for_status()  # Raises HTTPError if status code >= 400
    print("Response JSON:", bad_response.json())
except requests.exceptions.HTTPError as err:
    print("HTTP error occurred:", err)
except requests.exceptions.RequestException as err:
    print("Request failed:", err)
else:
    print("Request succeeded with status code:", bad_response.status_code)

# **Explanation**:
# - A 404 Not Found error will trigger an HTTPError.
# - `RequestException` is a catch-all for network-related errors.
````

## 7. Custom Headers

Some APIs require auth tokens or custom accept headers.


````python
headers = {
    'Authorization': 'Bearer YOUR_ACCESS_TOKEN',
    'Accept': 'application/json'
}

response = requests.get("https://api.github.com/user", headers=headers)
print("Status Code:", response.status_code)
print("Response JSON:", response.json())

# **Explanation**:
# - `headers` is a dictionary of header keys and values.
# - Common header fields include `Authorization`, `Content-Type`, `Accept`.
````

## 8. Using Sessions

`requests.Session()` lets you persist headers, cookies, and TCP connections across multiple requests.


````python
session = requests.Session()

# Example: update session-wide headers
session.headers.update({
    'User-Agent': 'MyApp/1.0'
})

response1 = session.get("https://api.github.com")
response2 = session.get("https://api.github.com/events")

print("Response 1 status:", response1.status_code)
print("Response 2 status:", response2.status_code)

# **Explanation**:
# - Sessions can improve performance by reusing connections.
# - Also useful for carrying authentication cookies between requests.
````

## 9. Real-World Examples with Concrete APIs

Below are some examples calling publicly available APIs that return meaningful data. No API keys required!

### 9.1 Pokémon API


````python
pokemon = "pikachu"
response = requests.get(f"https://pokeapi.co/api/v2/pokemon/{pokemon}")
if response.status_code == 200:
    data = response.json()
    print("Name:", data["name"])
    print("Base experience:", data["base_experience"])
    print("Types:", [t["type"]["name"] for t in data["types"]])
else:
    print("Error fetching data. Status code:", response.status_code)
````

**Explanation**:
- The [PokéAPI](https://pokeapi.co/) is a free API for Pokémon data.
- We retrieve info about \"pikachu\" and parse out its `name`, `base_experience`, and `types`.

### 9.2 Random Cat Fact

````python
response = requests.get("https://catfact.ninja/fact")
if response.status_code == 200:
    data = response.json()
    print("Random cat fact:", data["fact"])
else:
    print("Error:", response.status_code)
````

**Explanation**:
- This endpoint returns a random cat fact in JSON format.\n- `data[\"text\"]` gives the fact itself.

### 9.3 Bitcoin Price from Coindesk

````python
response = requests.get("https://data-api.coindesk.com//spot/v1/latest/tick?market=coinbase&instruments=BTC-USD&apply_mapping=true")
if response.status_code == 200:
    data = response.json()
    btc_usd_rate = data['Data']['BTC-USD']['PRICE']
    print("Current Bitcoin price in USD:", btc_usd_rate)
else:
    print("Error fetching Bitcoin price. Status code:", response.status_code)
````

**Explanation**:
- [Coindesk's public API](https://www.coindesk.com/coindesk-api/) provides current Bitcoin price data in various currencies.
- `data["bpi"]["USD"]["rate"]` returns a string representation of the USD price.
